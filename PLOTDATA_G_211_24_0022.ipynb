```python
import numpy as np
import matplotlib.pyplot as plt

# Contoh data (Volume vs waktu), silakan ganti dengan data asli Anda
t = np.array([1, 2, 3, 4, 5, 6])  # waktu
v = np.array([2.1, 4.1, 6.0, 8.1, 10.1, 12.2])  # volume

n = len(v)

# Hitung parameter rata-rata dan jumlah yang dibutuhkan
c1 = np.sum(t) / n       # rata-rata t
c2 = np.sum(v) / n       # rata-rata v
a = np.multiply(v, t)    # elemen v*t
c3 = np.sum(a)           # jumlah v*t
b = np.multiply(t, t)    # elemen t^2
c4 = np.sum(b)           # jumlah t^2

# Hitung parameter garis A dan B (v = At + B)
A = (c3 - n * c1 * c2) / (c4 - n * c1 * c1)
B = c2 - A * c1

# Plot garis regresi dengan titik-titik scatter
for tp in np.arange(t[0], t[-1], 2):
    vp = A * tp + B
    vp = vp * 100  # skala vp untuk plotting
    plt.scatter(tp, vp, color='g', s=1)

# Hitung galat RMS (Root Mean Square error)
sumee = 0
for i in range(len(t)):
    e = v[i] - (A * t[i] + B)
    ee = e * e
    sumee += ee

rms = np.sqrt(sumee / n)


plt.text(60, 28, 'v = At + B', color='g')
plt.arrow(78, 30, 6, 6, head_length=3, head_width=1.5, color='g', linewidth=0.5)


vp1 = A * t[0] + B
vp1 = '%.4f' % (vp1)
plt.text(2, 64, vp1, color='g')

vp2 = A * t[5] + B
vp2 = '%.4f' % (vp2)
plt.text(122, 25, vp2, color='g')


Ap = '%.5f' % (A)
plt.text(65, 18, 'A =', color='g')
plt.text(72, 18, Ap, color='g')


Bp = '%.5f' % (B)
plt.text(65, 12, 'B =', color='g')
plt.text(73, 12, Bp, color='g')


rmsp = '%.3f' % (rms)
plt.text(95, 3, 'RMS error =', color='g')
plt.text(123, 3, rmsp, color='g')


plt.show()
```

```python
"""
REGRESSION1
"""

# ----------------------- straight line fit to Volume v vs t -----------------------

n = len(v)

# Hitung nilai rata-rata dan jumlah
c1 = np.sum(t) / n           # rata-rata t
c2 = np.sum(v) / n           # rata-rata v
a = np.multiply(v, t)        # v * t (elemen per elemen)
c3 = np.sum(a)               # jumlah v * t
b = np.multiply(t, t)        # t^2 (elemen per elemen)
c4 = np.sum(b)               # jumlah t^2

# Hitung parameter garis regresi A dan B
A = (c3 - n * c1 * c2) / (c4 - n * c1 * c1)
B = c2 - A * c1

# Plot garis regresi sebagai titik-titik
for tp in np.arange(t[0], t[5], 2):
    vp = A * tp + B
    vp = vp * 100             # skala untuk plotting
    plt.scatter(tp, vp, color='g', s=1)

# ----------------------- Hitung galat RMS -----------------------
sumee = 0
for i in range(len(t)):
    e = v[i] - (A * t[i] + B)
    ee = e * e
    sumee += ee

rms = np.sqrt(sumee / n)
# ---------------------- labels ----------------------
plt.text(60, 28, 'v = At + B', color='g')
plt.arrow(78, 30, 6, 6, head_length=3, head_width=1.5, color='g', linewidth=0.5)

# ---------------------- beginning v value of line ----------------------
vp1 = A * t[0] + B
vp1 = '%.4f' % (vp1)  # reduce the number of decimal places
plt.text(2, 64, vp1, color='g')  # plot

# ---------------------- end v value of line ----------------------
vp2 = A * t[5] + B
vp2 = '%.4f' % (vp2)
plt.text(122, 25, vp2, color='g')

# ---------------------- print value of A ----------------------
Ap = '%.5f' % (A)
plt.text(65, 18, 'A =', color='g')
plt.text(72, 18, Ap, color='g')

# ---------------------- print value of B ----------------------
Bp = '%.5f' % (B)
plt.text(65, 12, 'B =', color='g')
plt.text(73, 12, Bp, color='g')

# ---------------------- print RMS error ----------------------
rms = '%.3f' % (rms)
plt.text(95, 3, 'RMS error =', color='g')
plt.text(123, 3, rms, color='g')

plt.show()
```
